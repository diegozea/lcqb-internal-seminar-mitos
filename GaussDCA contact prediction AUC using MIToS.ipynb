# Example: PDB 2IT0

````julia
pdbfile = downloadpdb("2IT0")
````

By default MIToS uses [PDBML](https://doi.org/10.1093/bioinformatics/bti082) files. These XML files are in general more curated. For example, chains and residues are renumbered using PDBe coordinates. They have `label_` and `auth_` numbers, e.g. [auth_comp_id](http://mmcif.rcsb.org/dictionaries/mmcif_pdbx_v50.dic/Items/_atom_site.auth_comp_id.html) and [](). We are going to use `auth_` by setting `label=false`.

```
<PDBx:atom_site id="1">  
         ...
         <PDBx:auth_asym_id>E</PDBx:auth_asym_id>
         <PDBx:label_asym_id>A</PDBx:label_asym_id>
         ...
</PDBx:atom_site>  
```

````julia
pdbresidues = read(pdbfile, PDBML, label=false)
````

````julia
write("2IT0.pdb", pdbresidues, PDBFile)
````

````julia
col2res = msacolumn2pdbresidue(msa, "IDER_MYCTU/3-62", "2IT0", "A", "PF01325", siftsfile, checkpdbname=true, missings=false)
````

````julia
resdict = @residuesdict pdbresidues model "1" chain "A" group "ATOM" residue All
````

````julia
[ three2residue(res.id.name) for (resnum, res) in resdict ]
````

````julia
cmap = msacontacts(msa, resdict, col2res)
````

````julia
?gaussdca
````

````julia
dca_scores = gaussdca(msa)
````

````julia
using ROCAnalysis
AUC(dca_scores, cmap)
````

````julia
function get_cmap(msa, seqname, pdb, chain)
    pdbfile = downloadpdb(pdb)
    pdbresidues = read(pdbfile, PDBML, label=false)
    col2res = msacolumn2pdbresidue(msa, seqname, pdb, chain)
    resdict = residuesdict(pdbresidues, "1", chain, "ATOM", All)
    msacontacts(msa, resdict, col2res)
end
````

````julia
cmap_two = get_cmap(msa, "IDER_MYCTU/3-62", "1U8R", "B")
````

````julia
AUC(dca_scores, cmap_two)
````
